```
import pandas as pd
```

```
import numpy as np
```

```
import matplotlib.pyplot as plt
```

```
%matplotlib inline
```

```
airline=pd.read_csv("airline_passengers.csv",index_col="Month")
```

```
airline.head(2)
```

```
airline.dropna(inplace=True)
```

```
airline.index=pd.to_datetime(airline.index)
```

```
airline['6-month-SWM']=airline['Thousands of Passengers'].rolling(window=6).mean()
```

```
airline['12-month-SWM']=airline['Thousands of Passengers'].rolling(window=12).mean()
```

```
airline.plot(figsize=(12,5))
```

```
airline['EWMA-12']=airline['Thousands of Passengers'].ewm(span=12).mean()
```

```
airline[['Thousands of Passengers','EWMA-12']].plot(figsize=(12,5))
```
